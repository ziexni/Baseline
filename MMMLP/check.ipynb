```python
video = pd.read_csv('datasets/video_feature.csv')
```

```python
video.head()
```

```python
video = video.sort_values('item_id')
```

```python
video['item_id'] = video['item_id'] - 1
```

```python
video.head()
```

```python
intr = pd.read_parquet('datasets/interaction.parquet')
```

```python
valid_item_ids = intr['item_id'].unique()  # 고유값 배열 반환

video_filtered = video[video['item_id'].isin(valid_item_ids)]  # 괄호 위치 수정
```

```python
item = item.merge(video_filtered, on='item_id', how='left')
```

```python
item
```

```python
item = item.drop(columns='index')
```

```python
item.to_parquet
```

```python
item = item.rename(columns={'video_feature_y':'video_feature'})
```

```python
item['video_feature'] = item['video_feature'].apply(
    lambda x: list(map(float, x.split(',')))
)
```

```python
print(type(item['video_feature'].iloc[0]))
```

```python
item.to_parquet('datasets/item_used.parquet', index=False)
```
